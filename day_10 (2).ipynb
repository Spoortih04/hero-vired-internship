```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\shekh\Documents\hero vired\spotify.csv")


#plt.style.use('dark_background')
genre_counts = df['Genre'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(genre_counts.index, genre_counts.values, color='black', alpha=0.7)

plt.title("Genre Counts")
plt.xlabel("Genre")
plt.ylabel("Counts")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales.csv')

# Histogram
plt.hist(df['Total_Sales'], color='skyblue', edgecolor='black')
plt.title('Total Sales Distribution')
plt.show()

# Bar graph
df.groupby('Product')['Total_Sales'].sum().plot.bar(color='orange')
plt.title('Sales by Product')
plt.show()

# Line plot
df.groupby('Date')['Total_Sales'].sum().plot.line(color='green')
plt.title('Sales Over Time')
plt.show()

# Scatter plot
plt.scatter(df['Quantity'], df['Total_Sales'], color='purple')
plt.title('Quantity vs Total Sales')
plt.xlabel('Quantity')
plt.ylabel('total_Sales')
plt.show()
```

```python
import pandas as pd

# Sample data: create a DataFrame with date and time strings
data = {
    'Order_ID': [1, 2, 3, 4],
    'Order_DateTime': [
        '2024-01-15 09:30:00',
        '2024-02-20 15:45:00',
        '2024-02-28 11:20:00',
        '2024-03-05 18:05:00'
    ],
    'Revenue': [100, 150, 200, 130]
}
df = pd.DataFrame(data)

# Convert 'Order_DateTime' to datetime format
df['Order_DateTime'] = pd.to_datetime(df['Order_DateTime'])

# Extract year, month, day, hour, minute, and second for further analysis
df['Year'] = df['Order_DateTime'].dt.year
df['Month'] = df['Order_DateTime'].dt.month
df['Day'] = df['Order_DateTime'].dt.day
df['Hour'] = df['Order_DateTime'].dt.hour
df['Minute'] = df['Order_DateTime'].dt.minute
df['Second'] = df['Order_DateTime'].dt.second


print(df)
```

```python
import pandas as pd

# Sample data: Revenue and Tax Rate
data = {
    'Order_ID': [1, 2, 3, 4],
    'Revenue': [100, 150, 200, 130],
    'Tax_Rate': [0.18, 0.18, 0.20, 0.18]
}
df = pd.DataFrame(data)

# Use a lambda function with apply to calculate revenue after tax
df['Revenue_After_Tax'] = df.apply(lambda row: row['Revenue'] * (1 + row['Tax_Rate']), axis=1)

# Use a lambda function to categorize orders as 'High' or 'Low' value
df['Value_Category'] = df['Revenue'].apply(lambda x: 'High' if x >= 150 else 'Low')

print(df)
```

```python
import pandas as pd
data = {
    'Name': ['spoorti','joshna','sam', 'apeksha', 'akhila', 'naina'],
    'Math': [50,49,45, 23, 67, 31],
    'Science': [48,39,50, 35, 70, 29],
    'English': [50,46,42, 30, 65, 40]
}

df = pd.DataFrame(data)

# Calculate total marks
df['Total'] = df[['Math', 'Science', 'English']].apply(lambda row: row.sum(), axis=1)

# Determine pass/fail
df['Result'] = df[['Math', 'Science', 'English']].apply(
    lambda row: 'Pass' if all(mark >= 35 for mark in row) else 'Fail', axis=1
)
print(df)
```
